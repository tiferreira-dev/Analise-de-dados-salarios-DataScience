```python
#Traduzindo o nome das colunas pra PT-BR
alterando_ptbr = ({"work_year": "ano",
    "experience_level": "nivel_experiencia",
    "employment_type": "contrato",
    "job_title": "cargo",
    "salary": "salario",
    "salary_currency": "moeda",
    "salary_in_usd": "usd",
    "employee_residence": "residencia",
    "remote_ratio": "remoto",
    "company_location": "empresa",
    "company_size": "tamanho_empresa",})

df.rename(columns=alterando_ptbr, inplace=True)
```

```python
#Observação da providência por tamanho da empresa
df['tamanho_empresa'].value_counts()
```

```python
#Traduzindo o nível de experiência
nivel_experiencia ={
    'SE' : 'Senior',
    'MI': 'Plêno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['nivel_experiencia'] = df['nivel_experiencia'].replace(nivel_experiencia)
df['nivel_experiencia'].value_counts()
```

```python
#Traduzindo o tipo de contrato
contrato = {
    'FT': 'Integral',
    'CT': 'Parcial',
    'PT': 'contrato',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
#Traduzindo o tamanho da empresa
tamanho_empresa = {
    'M': 'Grande',
    'L': 'Média',
    'S': 'Pequena'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
#Traduzindo o método de trabalho
mapa_trabalho = {
    0: 'Presencial',
    100:'Remoto',
    50: 'Híbrido'
}
df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['remoto'].value_counts()
```

```python
#Removendo valores nulos visto que pelo tamanho do DF esses valores são poucos e não vão alterar o resultado final
df_limpo = df.dropna()
df_limpo.isnull().sum()
```

```python
display(df_limpo)
```

```python
#Alterando o tipo do ano para int64
df_limpo = df_limpo.assign(ano = df_limpo['ano'].astype('int64'))
display(df_limpo)
```

```python
#Salvando os dados de limpeza do DF
df_limpo.to_csv('Limpeza de dados.cvs', index=False)
display(df_limpo)
```

```python
import plotly.express as px
#Qual é a média salarial por nível de senioridade?
#Calcular a média salarial em usd
media_nivel_experiencia = df_limpo.groupby('nivel_experiencia') ['usd'].mean().reset_index()

#Criar o gráfico
fig = px.bar(media_nivel_experiencia, x='nivel_experiencia', y='usd',
             title= 'Salário médio por nivel de experiência',
             labels= {'nivel_experiencia': 'experiência', 'usd': 'salário médios(USD)'},
             color='nivel_experiencia')

fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.show()
```

```python
#Qual é a proporção dos tipos de trabalho?
#Mostrando em gráfico donut char
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(
    remoto_contagem,
    names='tipo_trabalho',
    values='quantidade',
    title='Proporção dos Tipos de Trabalho',
    hole=0.5  #transforma em donut char
)
fig.update_traces(textinfo='percent+label')
fig.show()
     
```

```python
#Qual é o Top 5 de melhores médias salariais por cargo?
top_cargos = df_limpo.groupby('cargo')['usd'].mean().round(2).sort_values(ascending=False).head().reset_index()
top_cargos
```

```python
#Criação do gráfico
fig = px.bar(
    top_cargos,
    x='cargo',
    y='usd',
    title='Top 5 cargos com maiores médias salariais',
    labels={'usd': 'Salário médio (USD)', 'cargo': 'Cargo'}
)
fig.update_layout(xaxis_title='Cargo', yaxis_title='Salário médio (USD)')
fig.show()
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
